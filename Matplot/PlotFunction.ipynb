```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## sin(x),cos(x),tan(x),log(x)

```python
x = np.arange(0.1, 10, 0.1)# Set the range for x
```

```python
fig, axlist = plt.subplots(2,2)
axlist[0,0].plot(x,np.sin(x),color='r')
axlist[0,1].plot(x,np.cos(x),color='y')
axlist[1,0].plot(x,np.tanh(x),color='b')
axlist[1,1].plot(x,np.log(x),color='g')
plt.show()
```

## Sigmod function & tahn function

```python
def sigmoid(x):  
    return 1.0/(1+np.exp(-x))  
```

```python
def tahn(x):
    return 2/(1-np.exp(-2*x))-1
```

```python
x = np.linspace(-10, 10, 100)
plt.title("sigmoid and tahn")
plt.plot(x,sigmoid(x),label='sigmoid')
plt.plot(x,tahn(x),label = 'tahn')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
