## sin(x),cos(x),tan(x),log(x)

```python
x = np.arange(0.1, 10, 0.1)# Set the range for x
```

```python
fig, axlist = plt.subplots(2,2)
axlist[0,0].plot(x,np.sin(x),color='r')
axlist[0,1].plot(x,np.cos(x),color='y')
axlist[1,0].plot(x,np.tanh(x),color='b')
axlist[1,1].plot(x,np.log(x),color='g')
plt.show()
```

## Sigmod function & tahn function

```python
def sigmoid(x):  
    return 1.0/(1+np.exp(-x))  
```

```python
def tahn(x):
    return 2/(1-np.exp(-2*x))-1
```

```python
x = np.linspace(-10, 10, 100)
plt.title("sigmoid and tahn")
plt.plot(x,sigmoid(x),label='sigmoid')
plt.plot(x,tahn(x),label = 'tahn')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## plot_decision_boundary

```python
import numpy as np  
from sklearn.datasets import make_moons  
import matplotlib.pyplot as plt  
# generate random sactterplot
np.random.seed(0)  
X, y = make_moons(800, noise=0.20)  
```

```python
plt.scatter(X[:,0], X[:,1], s=20, c=y, cmap=plt.cm.Spectral)  
plt.show()  
```

```python
def plot_decision_boundary(pred_func):  
  
    # set the min and max  
    x_min, x_max = X[:, 0].min() - .3, X[:, 0].max() + .3 
    y_min, y_max = X[:, 1].min() - .3, X[:, 1].max() + .3
    h = 0.05  
  
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))  
  
    # prediction (classification) function
    Z = pred_func(np.c_[xx.ravel(), yy.ravel()])  
    Z = Z.reshape(xx.shape)  
  
    # Put the result into a color plot
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)  
    plt.axis('off')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral) 
```

```python
from sklearn.linear_model import LogisticRegressionCV  
clf = LogisticRegressionCV()  
clf.fit(X, y)  
       
# plot decision boundary
plot_decision_boundary(lambda x: clf.predict(x))  
plt.title("Logistic Regression")  
plt.show()  
```
